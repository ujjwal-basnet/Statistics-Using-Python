## using numpy 

```python
#sample datasets 
import numpy as np 
x = np.array([1  , 2 ,3 ,4 ,5 ])
y = np.array([2 , 4 ,1 , 3 ,5 ])
```

```python
#plotting scatter plot to see distributation 
import matplotlib.pyplot as plt
fig ,(ax1 ,ax2) = plt.subplots(1  , 2 , figsize = (12  , 5 ))

ax1.scatter(x  , y )
ax2.plot(x  , y  , marker = "."  , markersize = 40) 
plt.show()
```

```python
#varience is the measure of how far each datasets from mean 


def var_cal(x):
    var  = 0 
    for i in range(len(x)):
        difference  = x[i] - np.mean(x)
        difference = difference  * difference   
        var = var + difference
    return var / len(x)  # normalizing 
        
          
                
        
```

```python
# Covariance measures the extent to which two variables change together 
# it is calucated by  standard deviation  / mean 

def cov_cal (x , y ) : 
    mean_x = np.mean(x)
    mean_y = np.mean(y) 
    cov_XY = 0 
    if len(x) != len(y):
        raise ValueError("Arrays x and y must have the same length")
    
    for i in range(len(x)):
        spread_XY = (x[i] - mean_x)  * (y[i] - mean_y)
        cov_XY = cov_XY + spread_XY
            

    return cov_XY / len(x) #normalizing 

```
