```python
#data sets 
# X is increasing and y is also increasing 
X1 = [1,2,3,4,5]
y1 = [1,2,3,4,5]# if X is increasing and y is constant


X2 = [1,2,3,4,5]
y2 = [1,1,1,1,1]# if X is decreasing and y is increasing

X3 = [5,4,3,2,1]
y3 = [1,2,3,4,5]
```

```python
import matplotlib.pyplot as plt
fig, (ax1 , ax2 , ax3) = plt.subplots(1,3 , figsize = (15 , 4 ))
ax1.plot(X1  , y1)
ax2.plot(X2 , y2)
ax3.plot(X3 ,y3)


ax1.set_title('positively Corelated')
ax2.set_title('constant')
ax3.set_title('negatively corelated ')
plt.show()
```

### using numpy  


```python
import numpy as np 
print("corelation matrix\n " , np.corrcoef(X1 , y1))
print("\n" , np.corrcoef(X1 , y1)[0 , 1])
```

## using scipy

```python
#using scipy on negatively corelted datasets
from scipy.stats import pearsonr
r , p_value = pearsonr(X3 , y3 )
print("r is " , r )
print("p_value is "  ,   p_value)
```

# on pandas

```python
import pandas as pd
df = pd.DataFrame({'x' : X1 , 'y': y1})
print(df)
print("\n" , "corealation matrix is \n "  , df.corr())
```
